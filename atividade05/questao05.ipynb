```python
def potenciaDeBase(potencia, base):
  while potencia >= base:
    resto = potencia % base
    if resto != 0 :
      return False
    potencia = potencia // base

  if potencia == 1 :
    return True
```

```python
def validarPalavra(palavra, alfabeto):
  for letra in palavra:
    if letra not in alfabeto:
      return False
  return potenciaDeBase(len(palavra),2)
```

criando o Alfabeto da linguagem

```python
alfabeto = ["0"]
```

Testando algumas palavras

```python
print(validarPalavra("00", alfabeto))
print(validarPalavra("000", alfabeto))
print(validarPalavra("0000", alfabeto))
```
